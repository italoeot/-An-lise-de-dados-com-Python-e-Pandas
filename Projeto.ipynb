```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
#criando nosso dataframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#visualizando as 10 primeiras linhas
df.head(10)
```

```python
#quantidade dados em nossa base
df.shape
```

```python
#tipo de dados
df.dtypes
```

```python
#Qual a receita total?
df["Valor Venda"].sum()
```

```python
#Qual o custo total?
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"]) #criando nossa coluna de custo
```

```python
df.head(5)
```

```python
#Qual seria o custo total?
df["Custo"].sum()
```

```python
#Agora temos receita e custo. Como achar o lucro? (Receita - custo)
df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head(1)
```

```python
#Total Lucro
df["Lucro"].sum()
```

```python
#Vamos analisar agora o tempo de envio dos produtos
df["Tempo Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
#Vamos agrupar a média do tempo de envio por marcas
df.groupby("Marca")["Tempo Envio"].mean()
```

```python
#Vamos verificar valores perdidos / inconsistentes
df.isnull().sum()
```

```python
#Agrupando o lucro por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format #formatar nossos display númericos em float
```

```python
#resetando index (para formato colunar)
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Passando para análise gráfica
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
```

```python
#lucro por ano e análise em barras
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Lucro");
```

```python
#selecionando apenas vendas 2009
venda_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
venda_2009.head()
```

```python
#análise gráfica (2009)
venda_2009.groupby(df["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.xlabel("Lucro");
```

```python
#análise por marca (2009)
venda_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro");
```

```python
#análise por classe (2009)
venda_2009.groupby("Classe")["Lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro");
```
